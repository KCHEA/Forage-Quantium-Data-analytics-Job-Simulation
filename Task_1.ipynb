We will perform EDA, firstly on the purchase Dataframe
Since 'LYLTY_CARD_NBR' is just an ID, we will only focus on 'LIFESTAGE'	and 'PREMIUM_CUSTOMER'
Some interesting questions to ask will be:
1. What is the most and least common lifestage and premium?
2. What is the most and least common lifestage and premium combination?
3. How does lifestage affect premium?

```python
#List all the unique 'LIFESTAGE' and 'PREMIUM_CUSTOMER' values
display(purchase.groupby('LIFESTAGE').size().to_frame('LIFESTAGE COUNT').sort_values('LIFESTAGE COUNT', ascending=False))
print('\n\n')
display(purchase.groupby('PREMIUM_CUSTOMER').size().to_frame('PREMIUM_CUSTOMER COUNT').sort_values('PREMIUM_CUSTOMER COUNT', ascending=False))
print('\n\n')
display(purchase.groupby(['LIFESTAGE','PREMIUM_CUSTOMER']).size().to_frame('Count').sort_values('Count', ascending=False))
```

- Single people and couples are aggregated as one value
- Each category is differentiated based on age and if they have a families so the features will be determine by age (numerical values, eg. 0-99) and families(binary values, 0 or 1, yes or no)
- Premium can be list as Budget(lowest cost), Mainstream(mid cost), Premium(Highest cost)
- The trend is the older customers represent a larger portion of the customer base
